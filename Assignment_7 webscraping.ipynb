### Hent data ned om jobopslag med python og et andet valgfrit kodesprog

```python
import selenium_jobindex

python_entries = selenium_jobindex.get_info("Python")
print(python_entries)

print("========")

java_entries = selenium_jobindex.get_info("Java")
print(java_entries)
```

### Lav et plot der viser hvor mange jobopslag der er pr region mellem python og det andet kodesprog som blev valgt

```python
import matplotlib.pyplot as plt
import numpy as np

barwidth = 0.3
plt.figure()


# makes the x values for the bars
x_python = np.arange(len(python_entries.keys()))
x_java = [x + barwidth for x in x_python]


plt.title('Distribution of python and Java jobs by areas in Denmark', fontsize=12)

#sets the ticks on the x-axis and label the ticks. barwidth / 2 to set the tick between the bars
plt.xticks([r + (barwidth / 2) for r in range(len(python_entries.keys()))], python_entries.keys(), rotation = 90)


plt.xlabel("Area", fontsize=10)

plt.ylabel("Amount", fontsize=10)

y_plot = plt.bar(x_python, python_entries.values(), width=barwidth, align='center')
o_plot = plt.bar(x_java, java_entries.values(), width=barwidth, align='center')

plt.legend((y_plot, o_plot), ('Python', 'Java'))
plt.show()

```
